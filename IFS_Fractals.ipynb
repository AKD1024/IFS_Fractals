```python
import numpy as np
import random as rd

from matplotlib import pyplot as plt

Id = np.array([[1.,0.],[0.,1.]])

Its = 100000

def Sierp(x,y):
    out = np.array([[x,y]])
    old = np.array([x, y])
    for xx in range(Its):
        rand = rd.random()
        if rand < 1/3:
            new = 0.5*Id@old + 0.5*np.array([0.,0.])
        elif 1/3 <= rand < 2/3:
            new = 0.5*Id@old + 0.5*np.array([1.,0.])
        elif 2/3 <= rand < 1:
            new = 0.5*Id@old + 0.5*np.array([0.5,np.cos(np.pi/3)])
        out = np.append(out,[new],axis=0)
        old = new
    return out

Fract = Sierp(0,0)
    
plt.scatter(Fract[:,0],Fract[:,1],s=0.1)
```

```python
Its = 100000

def Fern(x,y):
    out = np.array([[x,y]])
    old = np.array([x, y])
    for xx in range(Its):
        rand = rd.random()
        if rand < 0.1:
            new = np.array([[0,0],[0,0.16]])@old
        elif 0.1 <= rand < 0.86:
            new = np.array([[0.85,0.04],[-0.04,0.85]])@old + np.array([0,1.6])
        elif 0.86 <= rand < 0.93:
            new = np.array([[0.2,-0.26],[0.23,0.22]])@old + np.array([0,1.6])
        elif 0.93 <= rand <= 1:
            new = np.array([[-0.15,0.28],[0.26,0.24]])@old + np.array([0,0.44])
        out = np.append(out,[new],axis=0)
        old = new
    return out

Fract = Fern(0,0)
    
plt.scatter(Fract[:,0],Fract[:,1],s=0.1)
```
